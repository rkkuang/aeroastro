```
\[Mu]1 = q/(1+q);
\[Mu]2 = 1/(1+q);
r0 = {X, 0};
x = {x1, x2};
r = {r1, r2};
lensequ = x - (r - \[Mu]1 (r-r0)/Dot[r-r0,r-r0] - \[Mu]2 (r+r0)/Dot[r+r0,r+r0])  (*=0*)
```
